```python
import os
import pandas as pd
import xarray as xr

# Define the directory containing the CSV files
directory = "."
output_file = "combined_data.nc"

# Initialize a list to store DataFrames
dfs = []

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        try:
            # Read the file into a Pandas DataFrame
            df = pd.read_csv(filepath)
            # Skip empty CSV files
            if not df.empty:
                dfs.append(df)
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# Combine all DataFrames into one
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    
    # Convert the combined DataFrame into an xarray Dataset
    dataset = xr.Dataset.from_dataframe(combined_df)
    
    # Save the Dataset to a NetCDF file
    dataset.to_netcdf(output_file)
    print(f"All non-empty CSVs combined and saved to {output_file}")
else:
    print("No non-empty CSV files found.")
```
