```python
import xarray as xr
import os
import pandas as pd

# Directory containing the .nc files
directory = '/data/esplab/aelyoussoufi/'

# List all .nc files in the directory that match the format 'swe.2003-12-01.2020-12-31.XX.nc'
files = [f for f in os.listdir(directory) if f.startswith('swe.2003-12-01.2020-12-31.') and f.endswith('.nc')]

# Loop through each file and process it
for file in files:
    # Open the NetCDF file using xarray
    ds = xr.open_dataset(os.path.join(directory, file))
    
    # Extract the snow_water_equivalent variable
    swe_data = ds['snow_water_equivalent'].values
    
    # Convert the xarray DataArray to a pandas DataFrame
    df = pd.DataFrame(swe_data.reshape(-1, swe_data.shape[-1]), columns=ds['X'].values)
    df['Y'] = ds['Y'].values
    
    # Add a column for the 'time' coordinate if needed
    df['time'] = ds['time'].values
    
    # Extract the state abbreviation from the filename (assumes format is correct)
    state_abbr = file.split('.')[-2]  # Extract the state abbreviation
    
    # Save the DataFrame to a CSV file
    csv_filename = os.path.join(directory, f"{state_abbr}_swe.csv")
    df.to_csv(csv_filename, index=False)
    print(f"Converted {file} to {csv_filename}")
```
