```python
import pandas as pd
import glob
import re

# Find all the CSV files matching our pattern
pattern = 'swe.2003-12-01.2020-12-31.*.csv'
csv_files = glob.glob(pattern)

def extract_state(filename):
    # Extract state abbreviation from filename
    match = re.search(r'\.([A-Z]{2})\.csv$', filename)
    return match.group(1) if match else None

# Create empty list to store dataframes
dfs = []

# Read each CSV and add a state column
for file in csv_files:
    try:
        # Read the CSV
        df = pd.read_csv(file)
        
        # Add state column
        state = extract_state(file)
        df['state'] = state
        
        # Append to our list
        dfs.append(df)
        print(f"Added {file} to combined dataset")
    except Exception as e:
        print(f"Error processing {file}: {str(e)}")

# Combine all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

# Save combined dataset
output_file = 'combined_swe_2003-2020.csv'
combined_df.to_csv(output_file, index=False)
print(f"\nSuccessfully created {output_file}")
print(f"Total rows: {len(combined_df)}")
```
