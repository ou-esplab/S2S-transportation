```python
import pandas as pd
import glob

# Get all CSV files starting with 'snow_wat'
csv_files = glob.glob("/data/esplab/aelyoussoufi/snow_wat*.csv")

# Combine all CSVs into a single DataFrame
combined_df = pd.concat((pd.read_csv(file) for file in csv_files), ignore_index=True)

# Save the combined DataFrame to a new CSV
combined_df.to_csv("../data/swe/csv/combined_snow_water.csv", index=False)

print("All files combined into 'combined_snow_water.csv'.")
```
