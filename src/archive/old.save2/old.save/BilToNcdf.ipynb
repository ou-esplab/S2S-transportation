```python
import os
import xarray as xr
import numpy as np

# Output directory for NetCDF files
output_dir = "./netcdf_output"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Grid details from metadata
ncols = 6935
nrows = 3351
ulxmap = -124.729583333333  # Upper left x-coordinate
ulymap = 52.8704166666666  # Upper left y-coordinate
xdim = 0.00833333333333333  # X resolution (degrees)
ydim = 0.00833333333333333  # Y resolution (degrees)

# Loop through all .bil files in the current directory
for filename in os.listdir("."):
    if filename.endswith(".bil"):
        filepath = os.path.join(".", filename)
        
        # Read the .bil file
        try:
            # Load data from .bil file as a NumPy array (16-bit signed integer)
            data = np.fromfile(filepath, dtype=np.int16)
            
            # Reshape to 2D array based on ncols and nrows
            data = data.reshape((nrows, ncols))
            
            # Create X and Y coordinates based on ulxmap, ulymap, and resolution
            x_coords = ulxmap + np.arange(ncols) * xdim
            y_coords = ulymap - np.arange(nrows) * ydim  # y decreases as we move down
            
            # Create an xarray Dataset
            ds = xr.Dataset(
                {
                    "Value": (["Y", "X"], data)
                },
                coords={
                    "X": x_coords,
                    "Y": y_coords,
                },
            )
            
            # Write to a NetCDF file
            output_filepath = os.path.join(output_dir, filename.replace(".bil", ".nc"))
            ds.to_netcdf(output_filepath)
            print(f"Converted {filename} to {output_filepath}")
        
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```
