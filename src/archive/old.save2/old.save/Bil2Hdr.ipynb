```python
import os
import numpy as np
import pandas as pd
```

```python
# Function to extract Year and Month from the filename
def extract_year_month(filename):
    # Example format: us_ssmv11034tS__T0001TTNATSYYYYMMDD05HP001.bil
    base_name = os.path.splitext(filename)[0]
    year = base_name[27:31]  # Extract YYYY
    month = base_name[31:33]  # Extract MM
    return year, month

# Loop through all .bil files in the current directory
for bil_file in os.listdir("."):
    if bil_file.endswith(".bil"):
        # Extract Year and Month from the filename
        year, month = extract_year_month(bil_file)

        # Read .bil file into a NumPy array (assuming binary float32 data for simplicity)
        data = np.fromfile(bil_file, dtype=np.int16)  # Adjust dtype based on your .bil format
        # Reshape the data based on known dimensions (e.g., ncols and nrows from the header)
        data = data.reshape(3351, 6935)  # Replace with actual nrows and ncols if different

        # Convert to DataFrame
        df = pd.DataFrame(data)
        # Append Year and Month columns
        df['Year'] = year
        df['Month'] = month

        # Save to CSV
        csv_file = f"{os.path.splitext(bil_file)[0]}.csv"
        df.to_csv(csv_file, index=False)
        print(f"Converted {bil_file} to {csv_file}")
```
