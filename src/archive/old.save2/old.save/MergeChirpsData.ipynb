```python
import xarray as xr
import glob

# List all the files in the current directory that match the pattern
files = sorted(glob.glob('chirps-v2.0.*.days_p25.nc'))

# Open all the files as a single dataset
ds = xr.open_mfdataset(files, combine='by_coords')

# Save the merged dataset to a new NetCDF file
ds.to_netcdf('merged_chirps_1981_2019.nc')

print("Files have been merged into 'merged_chirps_1981_2019.nc'")
```

```python
# Open the merged dataset
dt = xr.open_dataset('merged_chirps_1981_2019.nc')

# Extract the DJF months (December, January, February)
# Step 1: Add a 'month' coordinate to the dataset
dt['month'] = dt['time'].dt.month

# Step 2: Filter the dataset for months 12, 1, and 2
ds_djf = ds.where((dt['month'] == 12) | (dt['month'] == 1) | (dt['month'] == 2), drop=True)

# Remove the 'month' coordinate (optional, to keep the original structure)
ds_djf = ds_djf.drop_vars('month')

# Save the filtered dataset to a new NetCDF file
ds_djf.to_netcdf('merged_chirps_1981_2019_DJF.nc')

print("DJF months have been extracted and saved to 'merged_chirps_1981_2019_DJF.nc'")
```
