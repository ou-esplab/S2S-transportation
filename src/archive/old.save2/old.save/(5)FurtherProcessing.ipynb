```python
import pandas as pd
import glob
import re

# Find all the CSV files matching our pattern
pattern = 'swe.2003-12-01.2020-12-31.*.csv'
csv_files = glob.glob(pattern)

def extract_state(filename):
    # Extract state abbreviation from filename using the specific format
    match = re.search(r'swe\.2003-12-01\.2020-12-31\.([A-Z]{2})\.csv$', filename)
    return match.group(1) if match else None

# Create an empty list to store processed dataframes
processed_dfs = []

# Process each CSV file
for file in csv_files:
    try:
        # Read the CSV
        df = pd.read_csv(file)
        
        # Extract state code
        state = extract_state(file)
        if not state:
            print(f"Could not extract state from {file}")
            continue
            
        # Convert time column to datetime and extract year and month
        df['time'] = pd.to_datetime(df['time'])
        df['YEAR'] = df['time'].dt.year
        df['MONTH'] = df['time'].dt.month
        
        # Add state column
        df['swe_equivalent_data_state'] = state
        
        # Keep only the columns needed
        df = df[['YEAR', 'MONTH', 'lat', 'lon', 'snow_water_equivalent', 'swe_equivalent_data_state']]
        
        # Append processed dataframe to the list
        processed_dfs.append(df)
        print(f"Processed {file}")
        
    except Exception as e:
        print(f"Error processing {file}: {str(e)}")

# Combine all dataframes into a single dataframe
combined_df = pd.concat(processed_dfs, ignore_index=True)

# Sort by YEAR, MONTH, lat, and lon
combined_df = combined_df.sort_values(['YEAR', 'MONTH', 'lat', 'lon'])

# Save the combined dataset
output_file = 'combined_swe_2003-2020.csv'
combined_df.to_csv(output_file, index=False)

print(f"\nSuccessfully created {output_file}")
print(f"Total rows: {len(combined_df)}")
print("\nColumns in final dataset:")
print(combined_df.columns.tolist())
```
