```python
import xarray as xr
import pandas as pd
import glob
import os

# Find all matching NC files
pattern = 'swe.2003-12-01.2020-12-31.*.nc'
nc_files = glob.glob(pattern)

def convert_nc_to_csv(nc_file):
    # Read the NetCDF file
    ds = xr.open_dataset(nc_file)
    print(ds)
    # Convert to pandas DataFrame
    df = ds.to_dataframe()
    
    # Reset index to make all dimensions regular columns
    df = df.reset_index()
    
    # Simply replace .nc with .csv
    csv_file = nc_file.replace('.nc', '.csv')
    
    # Save to CSV
    df.to_csv(csv_file, index=False)
    
    # Close the dataset to free up memory
    ds.close()
    
    return csv_file

# Process each file
for nc_file in nc_files:
    try:
        csv_file = convert_nc_to_csv(nc_file)
        print(f"Converted: {nc_file} → {csv_file}")
    except Exception as e:
        print(f"Error converting {nc_file}: {str(e)}")
```
