```python
import pandas as pd
import os

# List all CSV files first to avoid repeated directory scanning
csv_files = [f for f in os.listdir(".") if f.endswith(".csv")]

# Process files in one go
for filename in csv_files:
    # Extract year and month directly
    year = filename[27:31]
    month = filename[31:33]
    
    # Read CSV and add columns in a single operation
    df = pd.read_csv(filename)
    df['Year'] = year
    df['Month'] = month
    df.to_csv(filename, index=False)
    
    print(f"Added Year ({year}) and Month ({month}) to {filename}")

print("Processing complete!")
```
